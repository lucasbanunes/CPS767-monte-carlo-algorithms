```python
import sympy
from sympy.utilities.lambdify import lambdify
import numpy as np
import matplotlib.pyplot as plt
```

```python
function_str = "max(10, 10*(1+sign(t-40)*exp(-(((t-300)/100)^2))))"
sympy_locals = dict(t=sympy.Symbol("t", real=True))
sym_func = sympy.sympify(function_str, locals=sympy_locals)
function_callable = lambdify(
    "t", sym_func, modules="numpy")
t = np.arange(0, 600, 0.01)
y = function_callable(t)

fig, ax = plt.subplots()
ax.grid(linestyle='--', alpha=0.5)
ax.set(
    title=function_str,
    xlabel="Time (s)",
    ylabel="Function Value",
)
ax.plot(t, y)
```
