```python
import sympy
from sympy.utilities.lambdify import lambdify
import numpy as np
import matplotlib.pyplot as plt
from utils import normal_distribution_str
```

```python
def get_arrival_rate_callable(function_str):
    sympy_locals = dict(t=sympy.Symbol("t", real=True))
    sym_func = sympy.sympify(function_str, locals=sympy_locals)
    function_callable = lambdify(
        "t", sym_func, modules="numpy")
    return function_callable
```

```python
mean = 300
std = 100
function_str = f"10*(1+250*{normal_distribution_str("t", mean, std)})"
t = np.arange(0, 600, 0.01)
arrival_rate_callable = get_arrival_rate_callable(function_str)
y = arrival_rate_callable(t)

fig, ax = plt.subplots()
ax.grid(linestyle='--', alpha=0.5)
ax.set(
    title=function_str,
    xlabel="Time (s)",
    ylabel="Function Value",
)
ax.plot(t, y)
print(function_str)
```
